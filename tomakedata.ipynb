```python
import pandas as pd
import numpy as np
from faker import Faker

# Initialize Faker for realistic names
fake = Faker()

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters
num_rows = 1000
salespeople = [fake.name() for _ in range(20)]  # 20 unique salespeople
products = ["Laptop", "Smartphone", "Headphones", "Monitor", "Keyboard", "Mouse", "Tablet"]
regions = ["North", "South", "East", "West"]
customer_types = ["New", "Returning"]

# Generate random data
data = {
    "sales_id": [f"TXN{str(i).zfill(5)}" for i in range(1, num_rows + 1)],
    "salesperson": np.random.choice(salespeople, num_rows),
    "date": pd.to_datetime("2023-01-01") + pd.to_timedelta(np.random.randint(0, 365, num_rows), unit="D"),
    "product": np.random.choice(products, num_rows),
    "quantity": np.random.randint(1, 10, num_rows),
    "unit_price": np.random.uniform(50, 1500, num_rows).round(2),
    "region": np.random.choice(regions, num_rows),
    "customer_type": np.random.choice(customer_types, num_rows)
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate total sales
df["total_sales"] = (df["quantity"] * df["unit_price"]).round(2)

# Display first few rows


```
